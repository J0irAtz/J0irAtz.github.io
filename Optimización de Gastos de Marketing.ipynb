````python
#Conteo de lanzamiento por a√±o
releases_by_year = (df.dropna(subset=['year_of_release'])
                      .groupby('year_of_release')['name']
                      .nunique())
display(releases_by_year.tail(20))  # √∫ltimos 20 a√±os para ver tendencia
releases_by_year.plot(kind='bar', figsize=(14,4), title='Juegos lanzados por a√±o')
````

````python
# 3.2 Plataformas con mayores ventas totales
platform_totals = (df.groupby('platform')['global_sales']
                     .sum()
                     .sort_values(ascending=False))
top_platforms = platform_totals.head(6).index.tolist()
display(platform_totals.head(10))

# Distribuci√≥n temporal de ventas para TOP plataformas
sales_year_platform = (df[df['platform'].isin(top_platforms)]
                        .dropna(subset=['year_of_release'])
                        .groupby(['year_of_release','platform'])['global_sales']
                        .sum()
                        .reset_index())

# Gr√°fico de l√≠neas por plataforma
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
for p in top_platforms:
    sub = sales_year_platform[sales_year_platform['platform']==p]
    plt.plot(sub['year_of_release'], sub['global_sales'], label=p)
plt.title('Ventas globales por a√±o - Plataformas TOP')
plt.xlabel('A√±o de lanzamiento')
plt.ylabel('Ventas (millones)')
plt.legend()
plt.show()

# Vida aproximada de cada plataforma (primer y √∫ltimo a√±o con ventas>0)
lifespan = (df[df['platform'].isin(top_platforms)]
            .dropna(subset=['year_of_release'])
            .groupby('platform')['year_of_release']
            .agg(['min','max','count'])
            .rename(columns={'count':'num_registros'}))
display(lifespan)
````

````python
# 3.3 Elegir periodo reciente (aj√∫stalo si tu gr√°fico sugiere otro mejor)
cut_year = 2013  # t√≠pico porque coincide con la generaci√≥n PS4/XOne; ajusta seg√∫n tus hallazgos
df_recent = df[df['year_of_release'].fillna(0).astype(int) >= cut_year].copy()

print("A√±os considerados:", sorted(df_recent['year_of_release'].dropna().unique()))
print("Filas en periodo reciente:", len(df_recent))
````

````python
import numpy as np

# Totales en el periodo reciente
leaders_recent = (df_recent.groupby('platform')['global_sales']
                  .sum()
                  .sort_values(ascending=False))
display(leaders_recent.head(10))

# Tendencia (pendiente) por plataforma usando regresi√≥n lineal simple
trend = []
for p, sub in df_recent.dropna(subset=['year_of_release']).groupby('platform'):
    yearly = sub.groupby('year_of_release')['global_sales'].sum()
    if len(yearly) >= 2:
        x = np.array(yearly.index, dtype=float)
        y = yearly.values
        slope = np.polyfit(x, y, 1)[0]  # pendiente: >0 crece, <0 cae
        trend.append((p, slope, yearly.sum()))
trend_df = (pd.DataFrame(trend, columns=['platform','slope','total_sales'])
              .sort_values(['slope','total_sales'], ascending=False))
display(trend_df.head(10))
display(trend_df.tail(10))
````

````python
# Boxplot para TOP plataformas recientes
top_recent = leaders_recent.head(6).index.tolist()
sub = df_recent[df_recent['platform'].isin(top_recent) & (df_recent['global_sales']>0)]

plt.figure(figsize=(10,6))
sub.boxplot(column='global_sales', by='platform')
plt.title('Distribuci√≥n de ventas globales por plataforma (periodo reciente)')
plt.suptitle('')
plt.xlabel('Plataforma')
plt.ylabel('Ventas (millones)')
plt.show()

# Estad√≠sticos descriptivos
display(sub.groupby('platform')['global_sales'].describe())
````

````python
# Juegos con m√∫ltiples plataformas en periodo reciente
multi = (df_recent.groupby('name')['platform']
         .nunique()
         .sort_values(ascending=False))

multi_names = multi[multi>=2].index

compare = (df_recent[df_recent['name'].isin(multi_names)]
           .groupby(['name','platform'])['global_sales']
           .sum()
           .reset_index())

# Muestra algunos ejemplos ordenados por ventas totales del juego
top_multi = (compare.groupby('name')['global_sales']
             .sum()
             .sort_values(ascending=False)
             .head(15)
             .index)

display(compare[compare['name'].isin(top_multi)]
        .sort_values(['name','global_sales'], ascending=[True,False]))
````

````python
from scipy import stats

# Filtrar datos para cada plataforma
xone_scores = df_recent[(df_recent['platform']=='XOne') & (~df_recent['user_score'].isna())]['user_score']
pc_scores   = df_recent[(df_recent['platform']=='PC') & (~df_recent['user_score'].isna())]['user_score']

# Prueba t de dos muestras (varianzas no iguales)
alpha = 0.05
stat, p = stats.ttest_ind(xone_scores, pc_scores, equal_var=False)

print("Xbox One vs PC")
print("p-valor:", p)
if p < alpha:
    print("Rechazamos H0 ‚Üí las medias son diferentes")
else:
    print("No se rechaza H0 ‚Üí no hay evidencia suficiente de diferencia")
````

````python
# Filtrar por g√©neros
action_scores = df_recent[(df_recent['genre']=='Action') & (~df_recent['user_score'].isna())]['user_score']
sports_scores = df_recent[(df_recent['genre']=='Sports') & (~df_recent['user_score'].isna())]['user_score']

# Prueba t
stat, p = stats.ttest_ind(action_scores, sports_scores, equal_var=False)

print("\nAcci√≥n vs Deportes")
print("p-valor:", p)
if p < alpha:
    print("Rechazamos H0 ‚Üí las medias son diferentes")
else:
    print("No se rechaza H0 ‚Üí no hay evidencia suficiente de diferencia")
````

Xbox One vs PC:

H‚ÇÄ: medias iguales.

H‚ÇÅ: medias diferentes.

Usamos una prueba t de dos muestras independientes porque comparamos promedios de dos grupos distintos.

Con un Œ± = 0.05:

Si p < 0.05 ‚Üí rechazamos H‚ÇÄ, hay diferencia significativa.

Si p ‚â• 0.05 ‚Üí no se rechaza H‚ÇÄ, no hay evidencia suficiente de diferencia.

Acci√≥n vs Deportes:

H‚ÇÄ: medias iguales.

H‚ÇÅ: medias diferentes.

Igual que antes, se aplica prueba t de dos muestras independientes.

Se interpreta el p-valor con respecto a Œ± = 0.05.
