# Paso 3: Construir el Modelo de Predicci√≥n

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler

# 1. Preparando los datos (X = caracter√≠sticas, y = objetivo)
X = df.drop('Churn', axis=1)
y = df['Churn']

# 2. Dividir en conjunto de entrenamiento y validaci√≥n
# Uso random_state=0 para que los resultados sean reproducibles
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)

# MODELO A: Regresi√≥n Log√≠stica
model_lr = LogisticRegression(random_state=0)
model_lr.fit(X_train_st, y_train)
predictions_lr = model_lr.predict(X_test_st)

# MODELO B: Bosque Aleatorio (Random Forest)
model_rf = RandomForestClassifier(n_estimators=100, random_state=0) 
# Nota: Los √°rboles no necesitan estandarizaci√≥n obligatoria, pero usaremos X_train para mantener la consistencia
model_rf.fit(X_train_st, y_train)
predictions_rf = model_rf.predict(X_test_st)

# Funci√≥n para imprimir m√©tricas
def print_metrics(y_true, y_pred, title):
    print(f"--- M√©tricas para {title} ---")
    print(f"Accuracy (Exactitud):  {accuracy_score(y_true, y_pred):.2f}")
    print(f"Precision:             {precision_score(y_true, y_pred):.2f}")
    print(f"Recall:                {recall_score(y_true, y_pred):.2f}")
    print("-" * 30)

# 3. Evaluando y comparando
print_metrics(y_test, predictions_lr, "Regresi√≥n Log√≠stica")
print_metrics(y_test, predictions_rf, "Bosque Aleatorio")
```

# Conclusion: 

Recall (Sensibilidad): 0.83 vs 0.81. Esto es lo m√°s cr√≠tico: la Regresi√≥n Log√≠stica detecta al 83% de los usuarios que se van a ir, mientras que el Bosque se pierde un poco m√°s.

Precisi√≥n: 0.85 vs 0.84. Tambi√©n es un poco m√°s precisa al dar la alerta.

Accuracy: Empate t√©cnico en 0.92.

Dado que nuestro objetivo de negocio es retener clientes, nos interesa el modelo que detecte a la mayor cantidad de "fugitivos" posible, as√≠ que la Regresi√≥n Log√≠stica es la ganadora.

# Paso 4: Crear Cl√∫steres de Usuarios

```python
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans

# 1. Estandarizaci√≥n de datos (Obligatorio para clustering)
# Uso todo el conjunto X (sin la columna Churn) porque quiero segmentar a TODOS los clientes
scaler = StandardScaler()
X_sc = scaler.fit_transform(X)

# A. DENDROGRAMA JER√ÅRQUICO
print("Generando dendrograma... (esto puede tardar unos segundos)")
linked = linkage(X_sc, method='ward')

plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top')
plt.title('Jerarqu√≠a de Cl√∫steres (Dendrograma)')
plt.show()

# B. MODELO K-MEANS (n=5)
# Las instrucciones piden expl√≠citamente 5 cl√∫steres
km = KMeans(n_clusters=5, random_state=0) # random_state para reproducibilidad
labels = km.fit_predict(X_sc)

# Guardo los cl√∫steres en el DataFrame original
df['cluster_km'] = labels

# C. AN√ÅLISIS DE LOS GRUPOS
# Ver el promedio de cada caracter√≠stica por cl√∫ster
cluster_means = df.groupby('cluster_km').mean()

# Agrego un conteo para saber qu√© tan grandes son los grupos
cluster_means['count'] = df.groupby('cluster_km')['cluster_km'].count()

print("\n--- Perfil Promedio de cada Cl√∫ster ---")
# Transpongo la tabla (.T) para que sea m√°s f√°cil de leer
display(cluster_means.T)

# Tasa de cancelaci√≥n por cl√∫ster
churn_rate = df.groupby('cluster_km')['Churn'].mean()
print("\n--- Tasa de Cancelaci√≥n (Churn Rate) por Cl√∫ster ---")
print(churn_rate.sort_values(ascending=False))
```

La segmentaci√≥n ha sido extremadamente reveladora. El algoritmo K-Means ha separado a los usuarios de una forma muy clara y l√≥gica para el negocio.

# Paso 5 Conclusiones y Recomendaciones 
# Estrategia de Retenci√≥n de Clientes

Tras analizar los datos de comportamiento de los usuarios de Model Fitness y aplicar un algoritmo de segmentaci√≥n (K-Means), hemos identificado patrones claros que distinguen a los clientes leales de aquellos en riesgo de cancelaci√≥n.

## 1. An√°lisis de Perfiles de Usuario (Clustering)

El algoritmo agrup√≥ a los usuarios en 5 categor√≠as distintivas seg√∫n su comportamiento y caracter√≠sticas:

* **Cluster 0 - "El Cliente Ideal" (Retenci√≥n Alta)**
    * **Tasa de Cancelaci√≥n:** 2.8% (La m√°s baja).
    * **Perfil:** Tienen contratos de largo plazo (promedio 11 meses). La mayor√≠a lleg√≥ a trav√©s de promociones de "amigos" o programas de empresas asociadas (Partners). Participan activamente en clases grupales.
    * **Conclusi√≥n:** La vinculaci√≥n social y contractual a largo plazo es la clave de su lealtad.

* **Cluster 4 - "El Atleta Constante" (Retenci√≥n Alta)**
    * **Tasa de Cancelaci√≥n:** 6.8%.
    * **Perfil:** Son los usuarios m√°s activos, con una frecuencia de visita de casi 3 veces por semana. Tienen la mayor antig√ºedad promedio (Lifetime) y son ligeramente mayores (promedio 30 a√±os).
    * **Conclusi√≥n:** El h√°bito consolidado de ejercicio es su principal motivo de permanencia.

* **Cluster 1 - "Los An√≥nimos" (Riesgo Moderado)**
    * **Tasa de Cancelaci√≥n:** 26.8%.
    * **Perfil:** Su caracter√≠stica definitoria es que **no proporcionaron n√∫mero de tel√©fono**. Tienen contratos de duraci√≥n media (4-5 meses).
    * **Conclusi√≥n:** La falta de datos de contacto limita nuestra capacidad para comunicarnos con ellos y fidelizarlos, aumentando el riesgo de fuga.

* **Cluster 2 - "Los Lejanos" (Riesgo Alto)**
    * **Tasa de Cancelaci√≥n:** 44.4%.
    * **Perfil:** Ninguno de estos usuarios vive o trabaja cerca del gimnasio (`Near_Location` = 0). Tienen contratos cortos y baja frecuencia de visitas.
    * **Conclusi√≥n:** La ubicaci√≥n geogr√°fica es una barrera f√≠sica insuperable para la retenci√≥n a largo plazo en este grupo.

* **Cluster 3 - "El Grupo Cr√≠tico" (Riesgo Muy Alto)**
    * **Tasa de Cancelaci√≥n:** 51.4% (M√°s de la mitad se va).
    * **Perfil:** A pesar de vivir cerca, tienen contratos mensuales (promedio 1.9 meses), asisten muy poco (1 vez/semana) y son los que menos participan en clases grupales.
    * **Conclusi√≥n:** Son usuarios "de prueba" que no logran generar el h√°bito ni integrarse en la comunidad del gimnasio.

## 2. Recomendaciones Estrat√©gicas

Basado en los perfiles anteriores, se proponen las siguientes acciones para reducir la tasa de cancelaci√≥n (churn):

### A. Incentivar el compromiso a largo plazo (Upselling)
* **Problema:** Los usuarios con contratos mensuales (Cluster 3) tienen un 51% de probabilidad de irse.
* **Acci√≥n:** Implementar ofertas agresivas para **pases de 6 y 12 meses** dirigidas espec√≠ficamente a usuarios nuevos en su primer mes. El objetivo es mover a los usuarios del perfil "Cluster 3" hacia el perfil "Cluster 0".

### B. Fomentar la integraci√≥n social y grupal
* **Problema:** Los grupos con mayor tasa de abandono son los que menos asisten a clases grupales.
* **Acci√≥n:** Incluir en el "paquete de bienvenida" pases gratuitos o exclusivos para clases grupales populares. Integrar al usuario en la comunidad desde la primera semana aumenta dr√°sticamente la retenci√≥n (como se ve en el Cluster 0).

### C. Optimizaci√≥n del Marketing Geolocalizado
* **Problema:** El Cluster 2 tiene una tasa de abandono del 44% debido a la lejan√≠a.
* **Acci√≥n:** Restringir la inversi√≥n publicitaria digital a c√≥digos postales estrictamente cercanos al gimnasio. Atraer clientes que viven lejos es ineficiente y genera una falsa sensaci√≥n de crecimiento, ya que terminan cancelando r√°pidamente.

### D. Campa√±a de Actualizaci√≥n de Datos
* **Problema:** El Cluster 1 (26% de churn) no tiene tel√©fono registrado.
* **Acci√≥n:** Lanzar una campa√±a interna (ej. "Actualiza tus datos y gana una bebida isot√≥nica") para capturar los n√∫meros de tel√©fono faltantes. Esto permitir√° incluirlos en campa√±as de SMS/WhatsApp de reactivaci√≥n y seguimiento.
